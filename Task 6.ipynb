```python
# IMPORTING REQUIRED LIBRARIES FOR THE DATASET
import pandas as pd
import sklearn.datasets as datasets
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier
```

```python
# Loading the iris dataset
iris=datasets.load_iris()
```

```python
# Forming the iris dataframe
df=pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size = 0.2, random_state = 6)
```

```python
#model
classifier=DecisionTreeClassifier()
classifier.fit(X_train, y_train)
print('Decision Tree Classifer Created')
```

```python
#visualizing the data
from sklearn.tree import plot_tree

fig = plt.figure(figsize = (15,15))
plot_tree(classifier, feature_names = iris.feature_names, class_names=iris.target_names, filled = True, rounded = True)
plt.show()
```

```python
#PREDICTING TRAINING SAMPLES
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import metrics
Y_pred = classifier.predict(X_train)

print("\nTraining score :")
print("Accuracy: ", metrics.accuracy_score(y_train, Y_pred))
#PREDICTING TESTING SAMPLES

Y_pred = classifier.predict(X_test)
print("\nTesting score :")
print("Accuracy: ", metrics.accuracy_score(y_test, Y_pred))
```
